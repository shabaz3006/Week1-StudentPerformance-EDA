```python
#step 1 load the dataset
import pandas as pd

data = {
    'gender': ['female', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'female', 'male'],
    'parental level of education': ['bcom', 'btech', 'bcom', 'btech', 'bcom', 'btech', 'bcom', 'btech', 'bcom', 'btech'],
    'lunch': ['standard', 'free/reduced', 'standard', 'standard', 'free/reduced',
              'standard', 'free/reduced', 'standard', 'standard', 'free/reduced'],
    'test preparation course': ['completed', 'none', 'none', 'completed', 'none',
                                'completed', 'none', 'completed', 'none', 'completed'],
    'math score': [85, 70, 90, 60, 75, 80, 55, 88, 67, 72],
    'reading score': [80, 68, 95, 62, 78, 84, 60, 90, 70, 75],
    'writing score': [82, 65, 93, 59, 76, 86, 58, 91, 68, 74]
}

df = pd.DataFrame(data)
df.to_csv("StudentsPerformance.csv", index=False)
df
```

```python
df.info()
df.describe()
df.columns
```

```python
#sep 5 check missing values
# Check for any missing or null values in each column
df.isnull().sum()
```

```python
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()
df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: use a clean style
sns.set(style="whitegrid")
sns.barplot(x='gender', y='math score', data=df)
plt.title('Average Math Score by Gender')
plt.show()
sns.boxplot(x='test preparation course', y='math score', data=df)
plt.title('Math Score Distribution by Test Prep')
plt.show()
plt.hist(df['math score'], bins=5, color='skyblue')
plt.title('Math Score Distribution')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.show()
```

```python
## Insights Summary

#1. Female students scored higher in reading and writing compared to males.
#2. Test preparation had a positive effect on scores across all subjects.
#3. Standard lunch students performed better than those with free/reduced lunch.
#4. Students with parents having bcom/btech background showed balanced performance.
#5. Reading and writing scores appear to be closely correlated.
```
